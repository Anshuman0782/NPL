```python
paragraph = "Ashutosh College is a college affiliated to the University of Calcutta , located in South Kolkata , adjacent to Gate 2 of Jatin Das Park Metro Station . The college was established in 1916 as South Suburban College under the chairmanship of educationist Sir Ashutosh Mukherjee , who was the Vice-Chancellor of the University of Calcutta . [ 1 ] After the death of Sir Ashutosh Mukherjee in 1924, the college was renamed Ashutosh College. Dr. Deepak Kumar Kar is the current principal of the college. The college is affiliated to the University of Calcutta "
```

```python
import nltk
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

```

```python
stemmer = PorterStemmer()
stemmer.stem("running")
```

```python
stemmer.stem("drinking")
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize('drinking')
```

```python

sentences = nltk.sent_tokenize(paragraph)
```

```python
print(sentences)
```

```python
import re
corpus = []
for i in range (len(sentences)):
    review = re.sub('[^a-zA-Z]', ' ', sentences[i])
    corpus.append(review.lower())
```

```python
corpus
```

```python
##Stemming
for i in corpus:
    words = nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(stemmer.stem(word))
    
```

```python
##Lemmatization
for i in corpus:
    words = nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(lemmatizer.lemmatize(word))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
X = cv.fit_transform(corpus)
```

```python
cv.vocabulary_
```

```python
## For Binary and N-gram Bag of Words
cv = CountVectorizer(binary=True, ngram_range=(2, 3))
X = cv.fit_transform(corpus)
X[0].toarray()
```

```python
cv.vocabulary_
```

```python
##TF-IDF Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(ngram_range=(2, 3),max_features=10)
X = cv.fit_transform(corpus)
```
